```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt






```

```python
#Dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ivan', 'Jack', 
             'Kelly', 'Liam', 'Mona', 'Nathan', 'Olivia', 'Paul', 'Quinn', 'Rose', 'Sam', 'Tina'],
    'Age': [25, 30, 45, 32, 28, 35, 40, 29, 31, 38, 26, 42, 33, 36, 27, 44, 30, 39, 24, 41],
    'Department': ['IT', 'HR', 'IT', 'Sales', 'HR', 'Sales', 'IT', 'Finance', 'Finance', 'Sales',
                   'IT', 'HR', 'Finance', 'IT', 'Sales', 'Finance', 'HR', 'Sales', 'IT', 'Finance'],
    'Salary': [55000, 48000, 95000, 62000, 51000, 68000, 88000, 72000, 75000, 64000,
               58000, 82000, 71000, 89000, 60000, 92000, 53000, 70000, 54000, 77000],
    'Experience': [2, 5, 20, 8, 4, 10, 15, 7, 9, 11, 3, 18, 8, 16, 5, 19, 6, 12, 1, 14],
    'PerformanceScore': [85, 78, 92, 81, 75, 88, 95, 84, 80, 82, 87, 90, 83, 94, 79, 91, 76, 89, 84, 86]
}

df = pd.DataFrame(data)
```

```python
#NumPy
salary_arr = df['Salary'].to_numpy()
exp_arr = df['Experience'].to_numpy()

mean_sal = np.mean(salary_arr)
median_sal = np.median(salary_arr)
std_sal = np.std(salary_arr)

max_exp = np.max(exp_arr)
min_exp = np.min(exp_arr)
```

```python
#Normalization
norm_salary = (salary_arr - np.min(salary_arr)) / (np.max(salary_arr) - np.min(salary_arr))

print("--- NumPy Analysis ---")
print(f"Salary - Mean: {mean_sal}, Median: {median_sal}, Std Dev: {std_sal:.2f}")
print(f"Experience - Max: {max_exp}, Min: {min_exp}\n")
```

```python
#Pandas
print("--- First 5 Rows ---")
print(df.head(), "\n")

print("--- Summary Statistics ---")
print(df.describe(), "\n")

avg_sal_dept = df.groupby('Department')['Salary'].mean()
best_employee = df.loc[df['PerformanceScore'].idxmax()]

df['Salary_After_Bonus'] = df['Salary'] * 1.10

print(f"Highest Performer: {best_employee['Name']} (Score: {best_employee['PerformanceScore']})\n")
```

```python
#Matplotlib
plt.figure(figsize=(12, 10))
```

```python
#BarChart
plt.subplot(2, 2, 1)
avg_sal_dept.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Salary by Department')
plt.ylabel('Salary ($)')
plt.xticks(rotation=45)

```

```python
#LinePlot
plt.subplot(2, 2, 2)
df_sorted = df.sort_values('Age')
plt.plot(df_sorted['Age'], df_sorted['Salary'], marker='o', linestyle='-', color='green')
plt.title('Age vs. Salary Trend')
plt.xlabel('Age')
plt.ylabel('Salary ($)')
plt.grid(True, alpha=0.3)
```

```python
#Piechart
plt.subplot(2, 2, 3)
df['Department'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, shadow=True)
plt.title('Department Distribution')
plt.ylabel('')
```

```python
#Histogram
plt.subplot(2, 2, 4)
plt.hist(df['Salary'], bins=6, color='orange', edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('Salary Range')
plt.ylabel('Number of Employees')

plt.tight_layout()
plt.show()
```
